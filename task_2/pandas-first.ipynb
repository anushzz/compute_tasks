```python
#prints top10 least danceable songs with their artists
df.sort_values('danceability')[['song_title','artist']].head(10)
```

```python
#prints 10 longest songs their artists and their durations
df.sort_values('duration_ms')[['song_title','artist','duration_ms']].tail(10)
```

```python
#prints song title artist duration and energy of top 10 highest energy songs whose duration is less than 400s 
df1=df[df['duration_ms']<400000]
df2=df1.sort_values('energy',ascending=False)[['song_title','artist','duration_ms','energy']].head(10)
df2
```

```python
#prints title and energy of songs whose eenrgy is >0.4
df3=df[df['energy']>0.4][['song_title','energy']]
df3
```

```python
df4=df[df['duration_ms']<300000]
df5=df4.sort_values('tempo',ascending=False).head(10)[['song_title','artist','duration_ms','tempo']]
df5
```

```python
df6=df[df['artist']=='Drake']
df7=df6[df6['energy']>0.5][['song_title','artist','energy']]
df7
```

```python
df8=df.sort_values('instrumentalness',ascending=False)[['song_title','artist','instrumentalness']].head(10)
df8
```

```python
x=df['artist'].value_counts()[:10].index.tolist()
avgdat=[]
for i in x:    
    avg=df[df['artist']==i]['tempo'].mean()
    avgdat.append({'artist': i, 'avg tempo': avg})
avgdf=pd.DataFrame(avgdat)
avgdf
    
```
